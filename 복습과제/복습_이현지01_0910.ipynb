## 정렬(내림차순, 오름차순) : **sort_values()**

```python
df.sort_values('english')
```

오름차순이 디폴트인 것 확인

```python
# 내림차순
df.sort_values('english', ascending=False)
```

```python
# 1. 'english'열 내림차순으로 정렬
# 2. 'english','id'열만 추출
# 3. english 값이 80 미만인 데이터 출력
df.sort_values('english', ascending=False)[['english', 'id']].query('english < 80')
```

### 두 개 이상의 조건으로 정렬 ([메인], [서브])

```python
# 'english': 메인, 'nclass': 서브
df.sort_values(['english', 'nclass'])
```

```python
# 'english'가 동일한 값인 경우, 'nclass'를 내림차순 정렬
df.sort_values(['english', 'nclass'], ascending=[True, False])
```

## 파생변수 만들기

1.   pandas에서 바로 작업 가능
2.   apply함수, lambda 적용
3.   loc, iloc 으로 생성
4.   assign() 함수 사용

```python
# pandas로 바로 생성
df['mean'] = (df['math'] + df['english'] + df['science']) / 3
```

```python
df
```

### assign() 함수

```python
# assign()
df.assign(mean_a = (df['math'] + df['english'] + df['science']) / 3)
```

한번에 여러가지 파생변수 생성 가능

```python
# head(5):위의 5개 행만 출력
df.assign(mean_mean = df['mean'].mean(), new = 0).head(5)
```

df['mean'].mean() : 개인 평균의 평균 'mean_mean' = 67.3

```python
import numpy as np
```

```python
# index값 출력 (위치)
np.where(df['mean'] > 67.3)
```

```python
# 평균 67.3 기준으로 평균 이상, 이하
# np.where() 에서 나온 index 위치에 값(평균이상,이하) 넣기
df.assign(result = np.where(df['mean'] > 67.3, '평균이상', '평균이하'))
```

### lambda

```python
df.assign(mean_l = lambda x : (x['math'] + x['english'] + x['science']) / 3 )
```

assign()으로 만든 변수는 따로 저장해야함 (자동 반영 X)

```python
df
```

## **groupby**

- df.groupby(기준).agg(변수, 요약통계치)



```python
# 'nclass'별로 group 나눈 후, 각 nclass별 math의 평균
df.groupby('nclass').agg(ncl_m_m = ('math', 'mean'))
```

```python
# 각 nclass별 math의 합계
df.groupby('nclass').agg(ncl_m_s = ('math', 'sum'))
```

```python
# 한번에 여러개 가능 (df_gp에 저장)
df_gp = df.groupby('nclass').agg(ncl_math_m = ('math', 'mean'),
                                 ncl_eng_m = ('english', 'mean'),
                                 ncl_eng_s = ('english', 'sum'))
```
