# **데이터 시각화**
- matplotlib, seaborn
- 데이터 분석에서의 가설을 설득하기 위한 과정

---



```python
from matplotlib import pyplot as plt
%matplotlib inline
```

## **matplot**
- figure: 도화지
- ax: 축
- 도화지에 큰 그림 그리기 -> 내부의 세부적인 부분 추가

### **matplotlib inline**

매직커맨드 show 하지 않고 바로 볼 수 있게 해줌

### 데이터 타입(연속형, 변주형) 에 따라 그릴 수 있는 시각화 달라짐

```python
plt.scatter([1,2,3,4],[5,6,7,8])
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

### 축 범위 설정 xlim, ylim

```python
plt.scatter([1,2,3,4],[5,6,7,8])
plt.xlim(0,3)
plt.ylim(5,7)
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

### 눈금값 설정 xticks, yticks

```python
plt.scatter([1,2,3,4],[5,6,7,8])
plt.xticks([1,2,3,4],['A','B','C','D'])
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

### 범례 설정 legend함수

```python
plt.scatter([1,2,3,4],[5,6,7,8], label='x=4')
plt.xticks([1,2,3,4],['A','B','C','D'])
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

### 추가 확장

```python
plt.scatter([1,2,3,4],[5,6,7,8], label='x=4')
plt.scatter([1,2,3,4],[8,7,6,5], label='x=3')
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

### 범례 위치 변경 가능
- left, center, right, upper left, lower left 등

```python
plt.scatter([1,2,3,4],[5,6,7,8], label='x=4')
plt.scatter([1,2,3,4],[8,7,6,5], label='x=3')
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc = 'center')
```

### 그래프의 크기 & 스타일 설정

```python
import seaborn as sns
sns.set(style = 'whitegrid')
plt.figure(figsize=(8,5))
plt.scatter([1,2,3,4],[5,6,7,8], label='x=4')
```

### 두 가지 이상의 그래프 그리기
## **plot** (선 그래프)

```python
plt.figure(figsize=(15,10))
plt.plot([1,2,3],[4,5,6], label= 'x1')
plt.plot([1,2,3],[6,5,4], label= 'x2')
plt.legend()
```

### 색상, 선 스타일, 두께 설정 가능
- color, linestyle, linewidth

```python
plt.figure(figsize=(15,10))
plt.plot([1,2,3],[4,5,6], color='red', linewidth=8, linestyle='dotted', label='x1')
plt.plot([1,2,3],[6,5,4], color='blue', linewidth=3, label='x2')
plt.legend()
```

### marker 설정
- . 점, o 원, s 사각형, * 별, D 다이아몬드 등
- marker, markersize 지정 가능

```python
plt.figure(figsize=(15,10))
plt.plot([1,2,3],[4,5,6], marker='*',markersize=20, color='red' ,label='x1')
plt.plot([1,2,3],[6,5,4], color='blue', label='x2')
plt.legend()
```

## **bar** (막대 그래프)

## **heatmap**
2차원 배열, 각 요소들을 색상으로 시각화
- annot = True : 데이터 표시
- fmt : 정수(d), 실수 표현
- cmap : 색상 조합

```python
import pandas as pd
df = pd.DataFrame({'+score':[5,4,3],
                  '-score':[2,1,5]},
                 index = ['Python','R','SQL'])
```

```python
sns.heatmap(df, annot=True, fmt='.1f', linewidth=2, linecolor='black')
```

```python
sns.heatmap(df, annot=True, fmt='.1f',cmap='Greys', linewidth=2, linecolor='black')
```

### 축 지정하여 실제 그래프 그리기

```python
x1 = np.arange(1,10)
x2 = np.arange(1,30)
y1 = x1*2
y2 = x2*3
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,7))

# 축 지정
ax[0].plot(x1,y1, color='red')
ax[1].bar(x2,y2, color='green')

ax[0].set_xlabel('ax[0] axis')
ax[1].set_xlabel('ax[1] axis')

ax[0].legend()
ax[1].legend()
```
