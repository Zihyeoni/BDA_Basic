### okt 이용해 동사, 명사 추출

```python
!pip install konlpy
```

```python
from konlpy.tag import Okt
```

```python
tw_tag = Okt()
```

```python
print('전체 형태소 추출 ', tw_tag.morphs(df.document[1]))
print('명사만 추출 ', tw_tag.nouns(df.document[1]))
print('태그 같이추출 ', tw_tag.pos(df.document[1]))
```

### 함수를 만들어 전체 데이터셋을 모두 전처리 진행해야함

```python
def my_tokenizer(doc):
    return[
        token
        for token, pos in tw_tag.pos(doc)
        if pos in ['Noun','Verb']
    ]

print('실제 결과 ', my_tokenizer(df.document[1]))
```

```python
re_cv = CountVectorizer(max_features=1000, tokenizer=my_tokenizer)
```

```python
re_dtm = re_cv.fit_transform(df_docu)
```

```python
for word, count in zip(re_cv.get_feature_names_out(), re_dtm[1].toarray()[0]):
    if count > 0:
        print(word, count, end=',')
```

### Cosine_Similarity 확인

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 기준으로 볼 리뷰
st = len(df_docu[0])//2
```

```python
source = df_docu[0][-st:]
```

```python
df_docu[0]
```

```python
st
```

```python
source
```

```python
# 코사인 유사도 게산하기 위해 벡터로 변환
source_rv=re_cv.transform([source])
```

```python
source_rv
```

```python
sim_res = cosine_similarity(source_rv, re_dtm)
```

```python
sim_res
```

```python
sim_res[0]
```

```python
print('유사도 정렬', sorted(sim_res[0], reverse=True)[:100])
```

```python
print('가장 유사한 리뷰는?', np.argmax(sim_res[0]))
```

## TF-IDF (Term Frequency - Inverse Document Frequency)
- 단어빈도 - 역문서빈도
- 카운트 대신 단어의 빈도에 그 단어가 출연한 문서의 수의 역수를 곱하는 것
- == 단어의 빈도를 그 단어가 나타난 문서의 수로 나누는 것
- 더 의미있는 단어 추출 가능

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
transformer = TfidfTransformer()
transformer
```

```python
rv_tfidf = transformer.fit_transform(re_dtm)
```

```python
rv_tfidf
```

```python
re_dtm[10].toarray()[0]
```

tf_idf (빈도 기반), 카운팅이 아닌 확률처럼 소수로 표현
