# **[필수 과제2] Melt 함수**
- weather file을 melt함수를 이용하여 정리하기



> image.png



```python
import pandas as pd
```

```python
weather = pd.read_csv('weather.csv')
```

```python
weather
```

```python
# d columns를 행으로 변환 (이름은 day로 지정
weather_d = pd.melt(weather, id_vars=['id', 'year', 'month', 'element'], var_name='day')
```

```python
weather_d
```

```python
# 결측치 제거
weather_d = weather_d.dropna()
```

```python
weather_d
```

```python
# month 기준으로 정렬
weather_d = weather_d.sort_values(by='month')
```

```python
weather_d
```

```python
# index 0부터 시작하도록 변경
weather_d = weather_d.reset_index()
```
