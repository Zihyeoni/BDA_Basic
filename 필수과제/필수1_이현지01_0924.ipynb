```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
```

```python
df = sns.load_dataset('mpg')
```

```python
df
```

```python
df.isna().sum()
```

```python
df_sp1 = df[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight']]
df_sp2 = df[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight']]
df_sp3 = df[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight']]
```

## simpleImputer로 결측치 채우기

### 1) strategy = **'mean'** (default, 평균값)

```python
imputer1 = SimpleImputer()
```

```python
df_mean = imputer1.fit_transform(df_sp1)
```

```python
# DataFrame으로 변환
df_mean = pd.DataFrame(df_mean)
```

```python
df_mean.isna().sum()
```

```python
df_mean.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight']
```

```python
df_mean
```

### 2) strategy = **'median'** (중앙값)

```python
imputer2 = SimpleImputer(strategy='median')
```

```python
df_median = imputer2.fit_transform(df_sp2)
```

```python
df_median = pd.DataFrame(df_median)
```

```python
df_median.isna().sum()
```

```python
df_median.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight']
```

```python
df_median
```

### 3) strategy = **'most_frequent'** (최빈값)

```python
imputer3 = SimpleImputer(strategy='most_frequent')
```

```python
df_mf = imputer3.fit_transform(df_sp3)
```

```python
df_mf = pd.DataFrame(df_mf)
```

```python
df_mf.isna().sum()
```

```python
df_mf.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight']
```

```python
df_mf
```

## 결과 비교

```python
print("horsepower 평균")
print("1. mean :", df_mean['horsepower'].mean())
print("2. median :", df_median['horsepower'].mean())
print("3. most_frequent :", df_mf['horsepower'].mean())
```

```python
print("horsepower 중앙값")
print("1. mean :", df_mean['horsepower'].median())
print("2. median :", df_median['horsepower'].median())
print("3. most_frequent :", df_mf['horsepower'].median())
```
