```python
from konlpy.tag import Okt
```

```python
k = Okt()
```

```python
# 명사
nouns = [n for w in words_filtered for n in k.nouns(w)]
print(nouns)
```

```python
# 동사
verbs = []
for w in words_filtered:
    tags = k.pos(w)
    verbs.extend([w for w, tag in tags if tag in ['Verb']])

print(verbs)
```

```python
# 토큰화 함수 정의
def tokenize(text):
    return k.nouns(text)

# 문서 토큰화하여 리스트에 저장
tokenized_documents = df['stripped_eliminated'].apply(tokenize)
```

```python
df['nouns_tokens'] = tokenized_documents
```

```python
# 'nouns_tokens' & 'label' 출력
print(df[['nouns_tokens', 'label']])
```
