```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

## 반복문 이용한 시각화 그래프

### barplot

```python
import matplotlib.gridspec as gridspec
grid = gridspec.GridSpec(2,2) # 4가지 그래프
```

```python
tt_features = ['sex', 'who', 'class', 'deck']
```

```python
plt.subplots_adjust(wspace=0.5, hspace=0.4)
plt.figure(figsize=(16,10))

for idx, feature in enumerate(tt_features):
    ax = plt.subplot(grid[idx])
    sns.barplot(x=feature, y='survived', data=df, ax=ax)
    ax.set_title(f'{feature} Distribution')
```

### countplot

```python
tt_columns = ['survived','sibsp','embarked']

fig, axs = plt.subplots(nrows=1, ncols=len(tt_columns), figsize=(15,3))

for idx, column in enumerate(tt_columns):
    sns.countplot(x=column, data=df, ax=axs[idx])
```
