```python
import pandas as pd
```

```python
df = pd.read_csv('spam(1).csv')
```

```python
df
```

```python
df['swap'] = df['text'].str.swapcase()
df
```

```python
df['swap'].str.lower()
```

```python
df['swap'].str.upper()
```

```python
df['swap'].str.capitalize()
```

```python
df['text'].str.center(width=50, fillchar='/')
```

```python
df['text'].str.split(n=1, expand=True)
```

```python
df['text'].str.replace('U', 'You', regex=False).str.replace('!', '!!!', regex=False).str.capitalize()
```

```python
df['text'].str.findall(pat='code').value_counts()
```

```python
df['num'] = df['text'].str.findall(pat='[0-9]+')
```
