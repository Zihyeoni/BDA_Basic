```python
import numpy as np
import pandas as pd
```

```python
import nltk
nltk.download('punkt')
nltk.download('webtext')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```python
df = pd.read_csv('movie_rv.csv')
```

```python
# 10,000개만 추출
df = df.iloc[:10000]
```

```python
df
```

```python
df1 = df['document']
```

```python
# 공백 제거
df1 = df1.str.strip()
```

```python
# 특수문자 제거
df1 = df1.str.replace(r'[^\uAC00-\uD7A30-9a-zA-Z\s]', '')
```

### Counter

```python
from collections import Counter
```

```python
from nltk.tokenize import word_tokenize
```

```python
w = ''
for i in df1[:]:
    w += i
    w += ' '
words = nltk.word_tokenize(w)
word_freq = Counter(words)
```

```python
word_freq
```
