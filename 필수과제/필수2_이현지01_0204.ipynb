```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('movie_rv.csv')
```

```python
df['document'] = df['document'].fillna(0)
df['document'] = df['document'].astype(str)
```

```python
# 공백 제거
df['new_doc'] = df['document'].str.strip().astype(str)
```

```python
# 특수문자 제거
df['new_doc'] = df['document'].str.replace(r'[^\uAC00-\uD7A30-9a-zA-Z\s]', '')
```

```python
df
```

# 1. 전처리 전

---



## 1) CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
re_cv = CountVectorizer(max_features=1000)
```

```python
re_dtm = re_cv.fit_transform(df['document'])
```

```python
re_dtm
```

```python
re_cv.get_feature_names_out()[:100]
```

```python
pip install konlpy
```

```python
from konlpy.tag import Okt
```

```python
tw_tag = Okt()
```

```python
def my_tokenizer(doc):
    return[
        token
        for token, pos in tw_tag.pos(doc)
        if pos in ['Noun','Verb']
    ]
```

```python
print('결과', my_tokenizer(df.document[5]))
```

```python
re_cv = CountVectorizer(max_features=1000, tokenizer=my_tokenizer)
```

```python
re_dtm = re_cv.fit_transform(df['document'])
```

```python
for word, count in zip(re_cv.get_feature_names_out(), re_dtm[5].toarray()[0]):
    if count > 0:
        print(word, count, end=',')
```

## 2) TF-IDF

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
transformer = TfidfTransformer()
transformer
```

```python
rv_tfidf = transformer.fit_transform(re_dtm)
```

```python
rv_tfidf
```

```python
tf = TfidfTransformer()
rv_tfidf = tf.fit_transform(re_dtm)
```

## 3) Cosine_Similarity 유사도 확인

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
st = len(df.document[5])//2
```

```python
source = df.document[5][-st:]
```

```python
df.document[5]
```

```python
st
```

```python
source
```

```python
source_cv = re_cv.transform([source])
```

```python
source_cv
```

```python
sim_res_cv = cosine_similarity(source_cv,re_dtm)
```

```python
sim_res_cv
```

```python
source_idf = tf.transform(source_cv)
```

```python
source_idf
```

```python
sim_res_tfidf = cosine_similarity(source_idf, rv_tfidf)
```

```python
sim_res_tfidf
```

```python
print('[CountVectorize] 가장 유사한 리뷰는?',np.argmax(sim_res_cv[0]))
```

```python
print('[TF-IDF] 가장 유사한 리뷰는?',np.argmax(sim_res_tfidf[0]))
```

```python
df.document[6728]
```

## 3) 차이점 분석

```python
# CountVectorizer 분석 결과
idx1 = list((-sim_res_cv[0]).argsort()[:10])
df.new_doc.loc[idx1]
```

```python
# TF-IDF 분석 결과
idx2 = list((-sim_res_tfidf[0]).argsort()[:10])
df.new_doc.loc[idx2]
```
